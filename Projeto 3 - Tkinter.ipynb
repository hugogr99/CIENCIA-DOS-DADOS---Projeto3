```python
from tkinter import *
import tkinter as tk
from PIL import ImageTk, Image
import os
from tkinter import messagebox

lista = []




class segunda_pagina(object):
    def __init__(self,janela2):
        self.frame1 = Frame(janela2)
        self.frame2 = Frame(janela2)
        self.frame3 = Frame(janela2)
        self.frame4 = Frame(janela2)
        
        self.bt_adicionar = Button(self.frame1,width=20,text='Criar item',command = self.adicionar_campo_de_entrada,bg='red',fg='white', activebackground='white', activeforeground='red')
        self.bt_adicionar.pack()
        
        self.bt_calcular = Button(self.frame3,width=20,text='PREVEJA',bg='green',fg='white',command=self.calcular)
        self.bt_calcular.pack(side=RIGHT)
        
        self.bt_ok = Button(self.frame3,width=20,text='Adicionar ao cálculo',bg='red',fg='white',command=self.bt_ok)
        self.bt_ok.pack(side=LEFT)
        
        self.frame1.pack(side=TOP,pady=20)
        self.frame2.pack()
        self.frame3.pack(side=BOTTOM,pady=20)
       
    
    def bt_ok(self):
        self.label['text'] = 'Adicionado'
        self.label['fg'] = 'green'
        dic = {}
        if self.variavel.get() == 'Ator/Atriz':
            dic['Ator'] = self.campo.get()
            lista.append(dic)
        if self.variavel.get() == 'Título':
            dic['Título'] = self.campo.get()
            lista.append(dic)
        if self.variavel.get() == 'Diretor(a)':
            dic['Diretor'] = self.campo.get()
            lista.append(dic)
        if self.variavel.get() == 'Produtor(a)':
            dic['Produtor'] = self.campo.get()
            lista.append(dic)

    def adicionar_campo_de_entrada(self):
        subframe = Frame(self.frame2)
        subframe.pack(side=TOP,pady=0)
        self.campo = Entry(subframe)
        #self.bt_ok = Button(subframe,width=5,text='Ok',bg='red',fg='white',command=self.bt_ok)
        self.label = Label(subframe,text='Não adicionado',fg='red')
        
        self.variavel = StringVar(subframe)
        self.variavel.set("Selecione")
        self.menu = OptionMenu(subframe,self.variavel,"Título","Ator/Atriz","Diretor(a)",'Roteirista','Produtor(a)')
        
        self.label.pack(side=RIGHT)
        self.campo.pack(side=LEFT)
        self.menu.pack(side=RIGHT)
        
        
    def calcular(self):
        self.inicia_janela3()
        
    def inicia_janela3(self):
        self.janela_3()
        
        
    def janela_3(self):
        resultado = 9
        self.janela3 = Toplevel()
        self.frame4 = Frame(self.janela3)
        self.frame5 = Frame(self.janela3)
        
        self.frame4.pack(side=TOP)
        self.frame5.pack()
        
        self.janela3.title('Previsor de sucesso de filme')
        self.janela3.wm_iconbitmap('icone_pipoca.ico')
        self.janela3.geometry('600x378')
        self.image = Image.open('cinema6.png')
        self.photo_image = ImageTk.PhotoImage(self.image)
        self.label = tk.Label(self.janela3, image = self.photo_image)
        self.label.place(x=0, y=0, relwidth=1, relheight=1)
        
        self.label_nota = Label(self.janela3,text = 'A nota do filme é')
        self.label_resultado = Label(self.janela3)
        self.label_status = Label(self.janela3)
        
        if resultado >= 7:
            self.label_resultado['bg'] = 'green'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Uau! Esse filme provavelmente é muito bom!'
            
        if resultado >= 5 and resultado < 7:
            self.label_resultado['bg'] = 'blue'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Pelo visto, o filme é mediano.'
            
        if resultado < 5:
            self.label_resultado['bg'] = 'red'
            self.label_resultado['fg'] = 'white'
            self.label_status['text'] = 'Que pena! Talvez o filme não seja muito bom...'
        
        self.label_resultado['text'] = str(resultado)
        self.label_nota.pack(side=TOP,pady=60)
        self.label_resultado.pack(side=TOP)
        self.label_status.pack(side=TOP,pady=60)
        
        
    
janela2 = Tk()

janela2.title('Previsor de sucesso de filme')
janela2.wm_iconbitmap('icone_pipoca.ico')
janela2.geometry('600x378')
image = Image.open('cinema6.png')
photo_image = ImageTk.PhotoImage(image)
label = tk.Label(janela2, image = photo_image)
label.place(x=0, y=0, relwidth=1, relheight=1)

segunda_pagina(janela2)

janela2.mainloop()
```
