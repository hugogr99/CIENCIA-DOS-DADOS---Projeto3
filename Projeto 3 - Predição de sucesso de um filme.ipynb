**Objetivo do projeto:**
A ideia principal do projeto, por ora, é criar um programa que leia informações sobre o filme como atores participantes, diretor, roteirista e produtor e de acordo com a probabilidade dos mesmos já terem atuado em filmes de sucesso, classificar o filme quanto à sua qualidade.
